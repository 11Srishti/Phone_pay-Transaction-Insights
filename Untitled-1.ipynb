```python
"""
Simplified HTTP utility with basic request support (standalone-safe version)
"""

import logging
import typing
import warnings
import sys

# --- Mock / fallback implementations (replace with real modules if available) ---
class HTTPResponse:
    def __init__(self, status=200, data=None):
        self.status = status
        self.data = data


class PoolManager:
    def request(
        self,
        method,
        url,
        body=None,
        fields=None,
        headers=None,
        preload_content=True,
        decode_content=True,
        redirect=True,
        retries=None,
        timeout=3,
        json=None,
    ):
        print(f"Making {method} request to {url}")
        return HTTPResponse(status=200, data="OK")


class Retry:
    pass


class Timeout:
    pass


# --- Logging setup ---
logger = logging.getLogger(__name__)
logger.addHandler(logging.NullHandler())


def add_stderr_logger(level: int = logging.DEBUG):
    """Add a stderr logger for debugging."""
    handler = logging.StreamHandler()
    handler.setFormatter(logging.Formatter("%(asctime)s %(levelname)s %(message)s"))
    logger.addHandler(handler)
    logger.setLevel(level)
    logger.debug("Logger initialized")
    return handler


# --- Warning helpers ---
def disable_warnings(category: type[Warning] = Warning):
    warnings.simplefilter("ignore", category)


# --- Global pool manager ---
_DEFAULT_POOL = PoolManager()


# --- Main request function ---
def request(
    method: str,
    url: str,
    *,
    body: typing.Any = None,
    fields: typing.Any = None,
    headers: typing.Mapping[str, str] | None = None,
    preload_content: bool = True,
    decode_content: bool = True,
    redirect: bool = True,
    retries: typing.Any = None,
    timeout: typing.Any = 3,
    json: typing.Any = None,
) -> HTTPResponse:
    """
    Simple HTTP request wrapper using a global PoolManager.
    """
    return _DEFAULT_POOL.request(
        method,
        url,
        body=body,
        fields=fields,
        headers=headers,
        preload_content=preload_content,
        decode_content=decode_content,
        redirect=redirect,
        retries=retries,
        timeout=timeout,
        json=json,
    )


# --- Example usage ---
if __name__ == "__main__":
    add_stderr_logger()

    response = request("GET", "https://example.com")
    print("Status:", response.status)
    print("Data:", response.data)
```
